Using the `penguins` dataset, make a scatter chart with `bill_length_mm` on the x-axis and `bill_depth_mm` on the y-axis.

```r
billchart = ggplot(penguins, aes(y=bill_depth_mm, x=bill_length_mm)) + geom_point() + labs(x='Bill Length', y= 'Bill Width')
billchart
```

Make the same plot but *encode* the `species` column with color.

```r
billchartcol = ggplot(penguins, aes(y=bill_depth_mm, x=bill_length_mm, color = species)) + geom_point() + labs(x='Bill Length', y= 'Bill Width', color = 'Species')
billchartcol
```
